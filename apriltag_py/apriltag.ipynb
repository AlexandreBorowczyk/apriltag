```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
cwd = os.getcwd()
```

```python
def imshow(image, figsize=(7, 7), cmap=None, title=''):
    plt.figure(figsize=figsize)
    plt.imshow(image, cmap=cmap)
    plt.title(title)
```

```python
image = cv2.imread(os.path.join(cwd, 'frame0000.jpg'), cv2.IMREAD_COLOR)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow(gray, cmap=plt.cm.gray, title='original image')
```

## Step 1 - Preprocessing

```python
sigma = 0.8
ksize = 3
gray_blur = cv2.GaussianBlur(gray, (ksize, ksize), sigma)
imshow(gray_blur, cmap=plt.cm.gray, title='blurred image')
```

```python
imshow((gray_blur - gray), cmap=plt.cm.gray)
```
